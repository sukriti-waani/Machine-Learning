## What is **Machine Learning** ?

Machine Learning (ML) is a field of Artificial Intelligence that enables computers to learn from data and improve their performance automatically, without being explicitly programmed.

Simple meaning:
Instead of writing rules manually, we give data to the machine, and it learns patterns from that data.

üìå Examples
Email spam detection, Movie recommendations (Netflix, YouTube), Traffic prediction in Google Maps, Face recognition

üîπ **Why Machine Learning is Needed?**
- Traditional programming fails for complex problems
- Huge amount of data is available
- Machines can learn patterns faster than humans
- Improves accuracy with more data

### **Types of Machine Learning**
Machine Learning is mainly divided into three types:
- Supervised Learning
- Unsupervised Learning
- Reinforcement Learning

Unsupervised Learning is a type of machine learning where the model is trained using unlabeled data, meaning no correct output is given.

‚öôÔ∏è How it Works
- Only input data is given

Reinforcement Learning is a type of machine learning where an agent learns by interacting with an environment and improves its actions using rewards and penalties.

‚öôÔ∏è How it Works

- Agent learns best action over time

üìå Examples
