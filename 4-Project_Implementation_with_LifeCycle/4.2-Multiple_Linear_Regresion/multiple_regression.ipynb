```python
# Lets prepare the dataframe

dataset = pd.DataFrame(california.data, columns=california.feature_names)
dataset.head()
```

```python
dataset['Price'] = california.target
```

```python
dataset.head()
```

```python
# Independent and Dependent features

# Select all columns except the last one as input features (Independent variables)
X = dataset.iloc[:, :-1]

# Select only the last column as output / target variable (Dependent variable)
y = dataset.iloc[:, -1]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=10)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

### Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression
```

```python
regression.fit(X_train, y_train)
```

```python
## slope or coefficients
regression.coef_
```

```python
# intercepts
regression.intercept_
```

```python
# Prediction for the test data

y_pred = regression.predict(X_test)
```

```python
y_test
```

```python
y_pred
```

```python
# Performance Metrics

from sklearn.metrics import mean_squared_error, mean_absolute_error

print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
## R square and adjusted R square

from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
print(score)
```

```python
# Adjusted R square
# display adjusted R-sqaured

1 - (1 - score) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
```

```python
# Assumptions

plt.scatter(y_test, y_pred)
plt.xlabel("Test Truth Data")
plt.ylabel("Test Predicted Data")
```

```python
residuals = y_test - y_pred
```

### **Pickling**

Python pickle module is used for serialising and de-serialising a Python object structure.
Any object in Python can be pickled so that it can be saved on disk. What pickle does is that it "serialises" the object forst before writing it to file. Pickling is a way to convert a python object (list, dict, etc.) into a character stream. The iea is that this character stream contains all the information necessary to reconstruct the object in another python script.

```python
import pickle

pickle.dump(regression, open('regressor.pkl', 'wb'))
```

```python
model = pickle.load(open('regressor.pkl', 'rb'))
```

```python
model.predict(X_test)
```
