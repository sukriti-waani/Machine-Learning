OVR is a technique used to apply **Logistic Regression** to problems
where the target has **more than two classes**.

Example classes:
- Cat  
- Dog  
- Horse


## 2Ô∏è‚É£ Why OVR is Required?

Logistic Regression works only for **binary classification**.

So for multiclass problems, we:
- Train multiple binary classifiers
- Each classifier separates **one class vs all other classes**


## 3Ô∏è‚É£ How OVR Works

If there are **K classes**, OVR creates **K models**.

For 3 classes ‚Üí 3 models:

Model 1 ‚Üí Class A vs Rest  
Model 2 ‚Üí Class B vs Rest  
Model 3 ‚Üí Class C vs Rest

For class A:

y = 1 ‚Üí if sample belongs to A  
y = 0 ‚Üí if sample belongs to any other class

## 8Ô∏è‚É£ Advantages of OVR

- Simple to implement  
- Works well for many real problems  
- Computationally efficient


## 9Ô∏è‚É£ Disadvantages of OVR

- Models may be imbalanced  
- Overlapping class probabilities  
- Less effective when classes are highly similar


## üîü Summary

- Logistic Regression is binary
