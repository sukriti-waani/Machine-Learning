#### 3Ô∏è‚É£ What is make_classification?

```python
from sklearn.datasets import make_classification
```

```python
# Create a fake (synthetic) classification dataset
X, y = make_classification(
    n_samples=1000,      # Total number of data records (1000 rows)
    n_features=10,       # Total number of input features (10 columns)
    n_classes=2,         # Number of output classes (Binary classification: 0 and 1)
    random_state=42     # Fixes random numbers so output is same every time
)
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,      # 30% of the data is kept for testing
    random_state=42     # Keeps the split same every time
)
```

```python
# Model training

from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression()
```

```python
# Train the Logistic Regression model using training data
# X_train contains the input features
# y_train contains the correct output labels
# The model learns the pattern between X_train and y_train
logistic.fit(X_train, y_train)
```

```python
# Use the trained Logistic Regression model to predict output
# for the testing input features X_test
y_pred = logistic.predict(X_test)

# Print the predicted class labels (0 or 1)
print(y_pred)
```

```python
# This function returns the probability of each class
# for every test data point in X_test

# The output will be in the form:
# [probability_of_class_0 , probability_of_class_1]

# Example output:
# [[0.20, 0.80],
#  [0.65, 0.35],
#  ...]

# This means:
# For first row ‚Üí 80% chance of class 1, 20% chance of class 0

probabilities = logistic.predict_proba(X_test)

# Print the probability values
print(probabilities)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Calculate accuracy of the model
# y_test  ‚Üí actual correct answers
# y_pred  ‚Üí answers predicted by the model
score = accuracy_score(y_test, y_pred)
print(score)

# Create confusion matrix
# It compares actual values (y_test) with predicted values (y_pred)
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Print detailed performance report of the model
# It shows precision, recall, f1-score and support
print(classification_report(y_test, y_pred))
```

## **Hyperparameter Tuning and Cross Validation**

```python
# Create a Logistic Regression model object
model = LogisticRegression()

# Different types of regularization techniques
# 'l1'  ‚Üí Lasso regularization (selects important features)
# 'l2'  ‚Üí Ridge regularization (reduces large coefficients)
# 'elasticnet' ‚Üí Combination of l1 and l2
penalty = ['l1', 'l2', 'elasticnet']

# Different values of C (regularization strength)
# C controls overfitting:
# Large C  ‚Üí less regularization
# Small C  ‚Üí more regularization
c_values = [100, 10, 1.0, 0.1, 0.01]

# Different optimization algorithms (solvers)
# newton-cg ‚Üí works with large datasets
# lbfgs     ‚Üí fast and commonly used
# liblinear ‚Üí good for small datasets
# sag       ‚Üí fast for large datasets
# saga      ‚Üí supports l1, l2 and elasticnet penalties
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
# Create a dictionary of all hyperparameters for Logistic Regression
# This will be used in GridSearchCV to try different combinations

params = dict(
    penalty = penalty,   # List of regularization techniques: l1, l2, elasticnet
    C = c_values,        # List of regularization strengths
    solver = solver      # List of optimization algorithms
)
```

```python
from sklearn.model_selection import StratifiedKFold

# Create a Stratified K-Fold cross-validation object
# It splits the data into K equal parts (folds)
# while keeping the same percentage of each class in every fold

cv = StratifiedKFold()
```

```python
# GridSearchCV is used to find the best hyperparameters automatically
from sklearn.model_selection import GridSearchCV

# Create GridSearchCV object
grid = GridSearchCV(
    estimator = model,     # Logistic Regression model
    param_grid = params,   # Dictionary of hyperparameters to test
    scoring = 'accuracy',  # Model is evaluated using accuracy
    cv = cv,               # Stratified K-Fold cross-validation
    n_jobs = -1            # Use all CPU cores for faster execution
)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
# Calculate accuracy of the model
# y_test  ‚Üí actual correct answers
# y_pred  ‚Üí answers predicted by the model
score = accuracy_score(y_test, y_pred)
print(score)

# Create confusion matrix
# It compares actual values (y_test) with predicted values (y_pred)
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Print detailed performance report of the model
# It shows precision, recall, f1-score and support
print(classification_report(y_test, y_pred))
```

```python
# roc_curve is used to draw ROC Curve
# roc_auc_score is used to calculate Area Under Curve (AUC)

from sklearn.datasets import make_classification   # Create fake classification dataset
from sklearn.linear_model import LogisticRegression # Logistic Regression model
from sklearn.model_selection import train_test_split # Split data into train & test
from sklearn.metrics import roc_curve   # Used to compute ROC curve values
from sklearn.metrics import roc_auc_score    # Used to calculate AUC score
from matplotlib import pyplot  # Used for plotting graphs
```

```python
# Generate a synthetic dataset for binary classification
X, y = make_classification(
    n_samples = 1000,     # Total number of data points
    n_classes = 2,        # Number of output classes (0 and 1)
    random_state = 1     # Fix randomness so data is same every time
)
```

```python
# Split the dataset into train and X_test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Generate a no-skill model prediction
# It always predicts class 0 for every test data point

dummy_model_prob = [0 for _ in range(len(y_test))]  # Create list of zeros equal to test data size

# Print dummy model predictions
dummy_model_prob
```

```python
# Lets create basic logistic model

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Predict probability of each class for the test dataset
# Output format ‚Üí [probability_of_class_0 , probability_of_class_1]
model_prob = model.predict_proba(X_test)
```

```python
# Select only the probability values of class 1
# model_prob[:, 1] means:
#   take all rows (:) and only column index 1 (class 1 probability)
model_prob = model_prob[:, 1]
```

```python
# lets calculate the score

dummy_model_auc = roc_auc_score(y_test, dummy_model_prob)
model_auc = roc_auc_score(y_test, model_prob)
print(dummy_model_auc)
print(model_auc)
```

# üìò Performance Metrics in Classification

---

## üî¥ False Positive Rate (FPR)

Definition:  
False Positive Rate shows how many negative cases are wrongly predicted as positive.

Formula:  
FPR = FP / (FP + TN)

Where:  
FP = False Positives  
TN = True Negatives  

---

## üü¢ Sensitivity

Definition:  
Sensitivity measures how well the model correctly identifies actual positive cases.

Formula:  
Sensitivity = TP / (TP + FN)

---

## üîµ Recall

Definition:  
Recall is the same as Sensitivity.

Formula:  
Recall = TP / (TP + FN)

---

## üü° Hit Rate

Definition:  
Hit Rate is another name for Sensitivity / Recall.

Formula:  
Hit Rate = TP / (TP + FN)

---

## üü£ True Positive Rate (TPR)

Definition:  
True Positive Rate shows the percentage of positives correctly predicted.

Formula:  
TPR = TP / (TP + FN)

---

## üìå Important Relationship

Sensitivity = Recall = Hit Rate = TPR

TPR = TP / P  = TP / (TP + FN)  = 1 - FNR

All these metrics have the same formula and meaning.

---

## üìù Exam Summary

| Metric | Formula |
|--------|--------|
| False Positive Rate (FPR) | FP / (FP + TN) |
| Sensitivity | TP / (TP + FN) |
| Recall | TP / (TP + FN) |
| Hit Rate | TP / (TP + FN) |
| True Positive Rate (TPR) | TP / (TP + FN) |

```python
# Generate ROC curve values for the dummy (no-skill) model
# y_test ‚Üí actual class labels
# dummy_model_prob ‚Üí probability scores predicted by dummy model
# dummy_fpr ‚Üí list of False Positive Rate values
# dummy_tpr ‚Üí list of True Positive Rate values
# _ ‚Üí threshold values (not used here)
dummy_fpr, dummy_tpr, _ = roc_curve(y_test, dummy_model_prob)

# Generate ROC curve values for the trained Logistic Regression model
# model_prob ‚Üí probability of class 1 predicted by trained model
# model_fpr ‚Üí list of False Positive Rate values
# model_tpr ‚Üí list of True Positive Rate values
# thresholds ‚Üí decision thresholds used to calculate FPR and TPR
model_fpr, model_tpr, thresholds = roc_curve(y_test, model_prob)
```

```python
thresholds
```

```python
model_fpr, model_tpr
```

```python
# Plot ROC curve of the dummy (no-skill) model
pyplot.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')

# Plot ROC curve of the trained Logistic Regression model
pyplot.plot(model_fpr, model_tpr, marker='.', label='Logistic')

# Label for X-axis ‚Üí False Positive Rate
pyplot.xlabel('False Positive Rate')

# Label for Y-axis ‚Üí True Positive Rate
pyplot.ylabel('True Positive Rate')

# Display legend to identify each curve
pyplot.legend()

# Show the final ROC curve plot
pyplot.show()
```

```python
import numpy as np   # Import numpy library for numerical operations

# Create a large figure for the plot
fig = pyplot.figure(figsize=(20, 50))

# Add a single subplot (one graph area) to the figure
ax = fig.add_subplot(111)

# Plot ROC curve for the dummy (no-skill) model using dashed line
ax.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')

# Plot ROC curve for the Logistic Regression model using dot markers
ax.plot(model_fpr, model_tpr, marker='.', label='Logistic')

# Loop through False Positive Rate, True Positive Rate and threshold values together
for fpr, tpr, thr in zip(model_fpr, model_tpr, thresholds):
    # Write the threshold value near each ROC point
    # np.round(thr, 2) rounds the threshold to 2 decimal places
    ax.annotate(str(np.round(thr, 2)), xy=(fpr, tpr))

# Set label for X-axis
ax.set_xlabel('False Positive Rate')

# Set label for Y-axis
ax.set_ylabel('True Positive Rate')

# Show legend to identify each ROC curve
ax.legend()

# Display the ROC curve plot
pyplot.show()
```
