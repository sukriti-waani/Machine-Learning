```python
# Define a function that takes processor name as input
def fetch_processor(text):
    # If the processor is one of the main Intel Core series
    # (i3, i5, or i7), return it as it is
    if text in ['Intel Core i7', 'Intel Core i5', 'Intel Core i3']:
        return text
    
    # If the processor starts with 'Intel' but is not i3/i5/i7
    # categorize it as 'Other Intel Processor'
    elif text.split()[0] == 'Intel':
        return 'Other Intel Processor'
    
    # Otherwise, it is assumed to be an AMD processor
    else:
        return 'AMD Processor'
```

```python
# Apply the function 'fetch_processor' on each value of the 'Cpu Name' column
# This converts detailed CPU names into categories like:
# 'Intel Core i7', 'Intel Core i5', 'Intel Core i3', 'Other Intel Processor', or 'AMD Processor'

df['Cpu brand'] = df['Cpu Name'].apply(fetch_processor)
```

```python
# Convert Memory column to string and remove ".0"
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)

# Remove 'GB' text
df['Memory'] = df['Memory'].str.replace('GB', '')

# Replace TB with 000 so that 1TB becomes 1000GB
df['Memory'] = df['Memory'].str.replace('TB', '000')

# Split memory column into two parts using '+'
new = df["Memory"].str.split("+", n=1, expand=True)

# Store first part
df["first"] = new[0]
df['first'] = df['first'].str.strip()

# Store second part
df['second'] = new[1]

# Identify storage type in first part
df['Layer1HDD'] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df['Layer1SSD'] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df['Layer1Hybrid'] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Extract only numeric values safely from first part
df['first'] = df['first'].str.extract('(\d+)')

# Fill missing second values with 0
df["second"].fillna("0", inplace=True)

# Identify storage type in second part
df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Extract only numeric values safely from second part
df['second'] = df['second'].str.extract('(\d+)')

# Replace NaN with 0 and convert both to integer
df[['first','second']] = df[['first','second']].fillna(0).astype(int)

# Calculate final storage values
df["HDD"] = (df["first"] * df["Layer1HDD"] + df["second"] * df["Layer2HDD"])
df["SSD"] = (df["first"] * df["Layer1SSD"] + df["second"] * df["Layer2SSD"])
df["Hybrid"] = (df["first"] * df["Layer1Hybrid"] + df["second"] * df["Layer2Hybrid"])
df["Flash_Storage"] = (df["first"] * df["Layer1Flash_Storage"] + df["second"] * df["Layer2Flash_Storage"])

# Drop temporary columns
df.drop(columns=['first','second','Layer1HDD','Layer1SSD','Layer1Hybrid',
                 'Layer1Flash_Storage','Layer2HDD','Layer2SSD',
                 'Layer2Hybrid','Layer2Flash_Storage'], inplace=True)
```

```python
# Define a function to categorize different operating system names
def cat_os(inp):
    
    # If OS is any version of Windows, return 'Windows'
    if inp == 'Windows 10' or inp == 'Windows 7' or inp == 'Windows 10 S':
        return 'Windows'
    
    # If OS is macOS or Mac OS X, return 'Mac'
    elif inp == 'macOS' or inp == 'Mac OS X':
        return 'Mac'
    
    # For all remaining OS types like Linux, No OS, or Others
    else:
        return 'Others/No Os/ Linux'
```

```python
df['os'] = df['OpSys'].apply(cat_os)
```

#### **Linear Regression**

```python
# Create a ColumnTransformer to apply OneHotEncoding only on selected categorical columns
# 'col_tnf' is the name given to this transformation step
# OneHotEncoder converts text categories into numerical dummy variables
# drop='first' removes one category to avoid multicollinearity (dummy variable trap)
# sparse=False returns a normal NumPy array instead of sparse matrix
# [0,1,7,10,11] are the column index positions that contain categorical values
# remainder='passthrough' keeps all other numerical columns unchanged

step1 = ColumnTransformer(transformers=[('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0,1,7,10,11])],remainder='passthrough')

# Create a Linear Regression model
step2 = LinearRegression()

# Build a pipeline that first preprocesses the data and then trains the model
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Train the entire pipeline on training data
pipe.fit(X_train, y_train)

# Predict laptop prices for test data
y_pred = pipe.predict(X_test)

# Print R2 score to evaluate how well the model explains the variance in price
print('R2 score', r2_score(y_test, y_pred))

# Print Mean Absolute Error to measure average prediction error
print('MAE', mean_absolute_error(y_test, y_pred))
```

#### **Voting Regressor**

```python
from sklearn.ensemble import VotingRegressor, StackingRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor
```

```python
# Apply OneHotEncoding to selected categorical columns and pass remaining numeric columns as it is
step1 = ColumnTransformer(
    transformers=[('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0,1,7,10,11])],
    remainder='passthrough'
)

# Random Forest Regressor model
rf = RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5,
                           max_features=0.75, max_depth=15, bootstrap=True)

# Gradient Boosting Regressor model
gbdt = GradientBoostingRegressor(n_estimators=100, max_features=0.5)

# XGBoost Regressor model
xgb = XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5)

# Extra Trees Regressor model
et = ExtraTreesRegressor(n_estimators=100, random_state=3, max_samples=0.5,
                         max_features=0.75, max_depth=15, bootstrap=True)

# Voting Regressor combining all models with weights
step2 = VotingRegressor(
    [('rf', rf), ('gbdt', gbdt), ('xgb', xgb), ('et', et)],
    weights=[5,1,1,1]
)

# Build final pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Train the voting regressor pipeline
pipe.fit(X_train, y_train)

# Predict on test data
y_pred = pipe.predict(X_test)

# Evaluate model performance
print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```

#### **Stacking**

```python
from sklearn.ensemble import StackingRegressor, VotingRegressor

# Apply OneHotEncoding to selected categorical columns and keep other columns unchanged
step1 = ColumnTransformer(
    transformers=[('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0,1,7,10,11])],
    remainder='passthrough'
)

# Define base models for stacking
estimators = [
    ('rf', RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5,
                                 max_features=0.75, max_depth=15, bootstrap=True)),
    ('gbdt', GradientBoostingRegressor(n_estimators=100, max_features=0.5)),
    ('xgb', XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5))
]

# Define stacking regressor with Ridge as final estimator
step2 = StackingRegressor(estimators=estimators, final_estimator=Ridge(alpha=100))

# Build pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Train the stacking regressor pipeline
pipe.fit(X_train, y_train)

# Predict laptop prices on test data
y_pred = pipe.predict(X_test)

# Evaluate performance using R2 score and Mean Absolute Error
print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```
