```python
# Split the 'ScreenResolution' column into two parts wherever 'x' occurs
# str.split('x')  splits string at x
# n=1  split only once
# expand=True  store result in two separate columns

# Example: "1920x1080 Touchscreen"  ["1920", "1080 Touchscreen"]

new = df['ScreenResolution'].str.split('x', n=1, expand=True)
```

```python
df['X_res'] = new[0]
df['Y_res'] = new[1]
```

```python
df.head()
```

```python
# Remove commas, extract numbers, and keep only the first number
df['X_res'] = df['X_res'].str.replace(',', '').str.findall(r'(\d+\.?\d+)').apply(lambda x: x[0])
# Remove commas like "1,920"  "1920"
# Extract numeric values using regex
# Take the first number from the list
```

```python
df.head()
```

```python
df.info()
```

```python
df['X_res'] = df['X_res'].astype('int')
df['Y_res'] = df['Y_res'].astype('int')
```

```python
df.info()
```

```python
# Calculate Pixels Per Inch (PPI) using screen resolution and screen size
# Formula: PPI = (X_res + Y_res)  Inches
# Convert the result to float type

df['ppi'] = (((df['X_res']**2) + (df['Y_res']**2))**0.5 / df['Inches']).astype('float')
```

```python
df.drop(columns=['ScreenResolution'], inplace=True)
```
