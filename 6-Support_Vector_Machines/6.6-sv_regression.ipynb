```python
# Independent and dependent features
X = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y = df['total_bill']
```

```python
# train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.head()
```

#### Feature Encoding (Label Encoding and Onehot Encoding)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex'] = le1.fit_transform(X_train['sex'])      
# Converts 'sex' column (text values like male/female) into numbers (0,1)

X_train['smoker'] = le2.fit_transform(X_train['smoker'])  
# Converts 'smoker' column into numeric form (0,1)

X_train['time'] = le3.fit_transform(X_train['time'])      
# Converts 'time' column into numbers (0,1)
```

```python
X_train.head()
```

```python
X_test['sex'] = le1.transform(X_test['sex'])      
# Converts 'sex' values in test data into numeric form using same mapping as training data

X_test['smoker'] = le2.transform(X_test['smoker'])  
# Converts 'smoker' values using encoding learned from training data

X_test['time'] = le3.transform(X_test['time'])      
# Converts 'time' values using same encoding as training data
```

```python
X_test.head()
```

```python
# Onehot Encoding ---- ColumnTransformer

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(
    transformers=[('onehot', OneHotEncoder(drop='first'), [3])],
    remainder='passthrough'
)
# ColumnTransformer → applies different transformations to different columns

# ('onehot', OneHotEncoder(drop='first'), 3)
# onehot → name of the transformer
# OneHotEncoder(drop='first') → converts categorical column into dummy variables
#                                drop='first' removes first category to avoid dummy trap
# 3 → column index to apply one-hot encoding on (4th column)

# remainder='passthrough' → all other columns remain unchanged
```

```python
import sys
import numpy

numpy.set_printoptions(threshold=sys.maxsize)
# Allows NumPy to display all values of large arrays without truncating

ct.fit_transform(X_train)
# fit_transform() learns the encoding rules from X_train
# and then transforms X_train using ColumnTransformer ct
# Output is the fully encoded training dataset
```

```python
X_test = ct.transform(X_test)
```

```python
X_test
```

```python
# SVR = Support vector Regression

from sklearn.svm import SVR
svr = SVR()
```

```python
svr.fit(X_train, y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```

```python
## Hyperparameter tuning using GridSearch CV

from sklearn.model_selection import GridSearchCV

# GridSearchCV is used to find the best combination of hyperparameters

# Defining the range of parameters to test
param_grid = {

    'C' : [0.1, 1, 10, 100, 1000],
    # C → Regularization parameter
    # Controls how much the model tries to avoid misclassification
    # Small C  → allows more mistakes (simple model)
    # Large C  → tries to classify all points correctly (complex model)

    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    # gamma → controls how far the influence of a point reaches
    # High gamma → decision boundary is very tight
    # Low gamma  → decision boundary is smooth and wide

    'kernel': ['rbf']
    # kernel is fixed as RBF (non-linear kernel)
}
```

```python

grid = GridSearchCV(SVR(), param_grid=param_grid, refit=True, cv=5, verbose=3)

# GridSearchCV() → automatically finds the best hyperparameters for SVC model

# SVC() → base Support Vector Classifier model

# param_grid = param_grid → parameters list (C, gamma, kernel) that we defined earlier

# refit = True → after finding the best parameters, model is trained again on full dataset

# cv = 5 → uses 5-fold cross-validation
# Data is split into 5 parts
# Model trains on 4 parts and tests on 1 part (repeated 5 times)

# verbose = 3 → shows detailed progress of training on the screen
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_prediction = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(r2_score(y_test, grid_prediction))
print(mean_absolute_error(y_test, grid_prediction))
```
