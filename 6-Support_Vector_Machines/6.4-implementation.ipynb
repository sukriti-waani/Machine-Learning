```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
# Create synthetic data points

from sklearn.datasets import make_classification
```

```python
from sklearn.datasets import make_classification

# make_classification() is used to create a synthetic classification dataset

X, y = make_classification(
    n_samples=1000,        # Total number of data points (rows) = 1000

    n_features=2,         # Total number of input features (columns) = 2 
    # So each data point has 2 input values (X1, X2)

    n_classes=2,          # Number of output classes = 2
    # Example: 0 and 1 (binary classification)

    n_clusters_per_class=1, # Each class will have only 1 cluster
    # So class 0 forms one group and class 1 forms one group

    n_redundant=0         # No redundant features
    # Means no feature is copied or dependent on another feature
)

# X → Feature matrix (1000 rows × 2 columns)
# y → Target labels array (1000 values of 0 or 1)
```

```python
pd.DataFrame(X)[0]   # Converts X (array) into a DataFrame and selects the first column

# pd.DataFrame(X) → converts the feature matrix X into a table format (DataFrame)
# [0] → selects the column with index 0 (first feature)
# So this returns all values of the first feature of X
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)

# sns.scatterplot() is used to draw a scatter plot (points on a graph)

# x = pd.DataFrame(X)[0] → selects the first feature (X-axis values)

# y = pd.DataFrame(X)[1] → selects the second feature (Y-axis values)

# hue = y → colors the points based on class labels in y
#            Points with class 0 get one color
#            Points with class 1 get another color

# This plot helps us visually see how the two classes are distributed
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel = 'linear')
```

```python
svc.fit(X_train, y_train)
```

```python
svc.coef_
```

```python
# Prediction

y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
rbf = SVC(kernel='rbf')   # Creates a Support Vector Classifier with RBF kernel

# SVC → Support Vector Classifier (used for classification problems)

# kernel = 'rbf' → Radial Basis Function kernel
# This kernel transforms data into higher dimension
# It helps separate non-linear data (when classes are not in straight line)

# rbf model is now ready to be trained using rbf.fit(X_train, y_train)
```

```python
rbf.fit(X_train, y_train)
```

```python
# Prediction
y_pred1 = rbf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred1))
print(confusion_matrix(y_test, y_pred1))
```

```python
polynomial = SVC(kernel='poly') 
polynomial.fit(X_train, y_train)
# Prediction
y_pred2 = polynomial.predict(X_test)
print(classification_report(y_test, y_pred2))
print(confusion_matrix(y_test, y_pred2))
```

```python
sigmoid = SVC(kernel='sigmoid') 
sigmoid.fit(X_train, y_train)
# Prediction
y_pred3 = sigmoid.predict(X_test)
print(classification_report(y_test, y_pred3))
print(confusion_matrix(y_test, y_pred3))
```

```python
sigmoid.intercept_
```

### Hyperparameter Tuning with SVC

```python
from sklearn.model_selection import GridSearchCV

# GridSearchCV is used to find the best combination of hyperparameters

# Defining the range of parameters to test
param_grid = {

    'C' : [0.1, 1, 10, 100, 1000],
    # C → Regularization parameter
    # Controls how much the model tries to avoid misclassification
    # Small C  → allows more mistakes (simple model)
    # Large C  → tries to classify all points correctly (complex model)

    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    # gamma → controls how far the influence of a point reaches
    # High gamma → decision boundary is very tight
    # Low gamma  → decision boundary is smooth and wide

    'kernel': ['rbf']
    # kernel is fixed as RBF (non-linear kernel)
}
```

```python
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(SVC(), param_grid=param_grid, refit=True, cv=5, verbose=3)

# GridSearchCV() → automatically finds the best hyperparameters for SVC model

# SVC() → base Support Vector Classifier model

# param_grid = param_grid → parameters list (C, gamma, kernel) that we defined earlier

# refit = True → after finding the best parameters, model is trained again on full dataset

# cv = 5 → uses 5-fold cross-validation
# Data is split into 5 parts
# Model trains on 4 parts and tests on 1 part (repeated 5 times)

# verbose = 3 → shows detailed progress of training on the screen
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
# Prediction
y_pred4 = grid.predict(X_test)
print(classification_report(y_test, y_pred4))
print(confusion_matrix(y_test, y_pred4))
```
