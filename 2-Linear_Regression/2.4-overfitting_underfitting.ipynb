(Bias and Variance Trade-off)


## 1Ô∏è‚É£ Dataset Split

In Machine Learning, data is divided into:

1. Training Dataset  
2. Validation Dataset  
3. Test Dataset


### Typical Split

- Training data ‚Üí 70%
- Test data ‚Üí 30%

Example:
If total data points = 1000  
Training data = 700  
Test data = 300


## 2Ô∏è‚É£ Purpose of Each Dataset

Training Dataset:
- Used to train the model
- Model learns patterns from this data

Validation Dataset:
- Used for hyperparameter tuning
- Helps improve model performance

Test Dataset:
- Used only for final evaluation
- Checks how well model generalizes


## 3Ô∏è‚É£ Bias and Variance

Bias:
- Error due to oversimplified model
- Model fails to capture pattern

Variance:
- Error due to too complex model
- Model fits noise in data


## 4Ô∏è‚É£ Underfitting

Underfitting happens when:
- Model is too simple
- Cannot learn the data pattern properly


### Characteristics of Underfitting

- Training accuracy ‚Üí Low
- Test accuracy ‚Üí Low
- High Bias
- Low Variance

### Conclusion

If both training and test accuracy are low,
the model is **Underfitting**.



## 5Ô∏è‚É£ Overfitting

Overfitting happens when:
- Model is too complex
- Learns noise along with data


### Characteristics of Overfitting

- Training accuracy ‚Üí Very High
- Test accuracy ‚Üí Low
- Low Bias
- High Variance


### Conclusion

If training accuracy is high but test accuracy is low,
the model is **Overfitting**.


## 6Ô∏è‚É£ Generalized (Ideal) Model

A good model should:
- Perform well on training data
- Perform well on test data


### Characteristics of Generalized Model

- Training accuracy ‚Üí High
- Test accuracy ‚Üí High
- Low Bias
- Low Variance


## 7Ô∏è‚É£ Summary Table

Underfitting:
- Low train accuracy
- Low test accuracy
- High bias, Low variance

Overfitting:
- High train accuracy
- Low test accuracy
- Low bias, High variance

Generalized Model:
- High train accuracy
- High test accuracy
- Low bias, Low variance


## 8Ô∏è‚É£ Summary

- Underfitting ‚Üí model too simple
- Overfitting ‚Üí model too complex
- Bias measures error from simplicity
- Variance measures error from complexity
- Goal is to achieve good generalization
