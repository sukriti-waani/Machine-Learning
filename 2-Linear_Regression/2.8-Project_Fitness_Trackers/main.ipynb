```python
# Randomly select 80% of the dataset to create training data
# frac=0.8 means 80% rows are selected
# random_state=42 ensures the same rows are selected every time
train_df = df.sample(frac=0.8, random_state=42)
```

```python
# Create a figure with 2 rows and 3 columns of subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(14, 10))

# Loop through each numerical column with its index number
for i, column in enumerate(numerical_data):

    # Decide row position of subplot
    # Example: 0//3 = 0, 3//3 = 1
    row = i // 3

    # Decide column position of subplot
    # Example: 0%3 = 0, 1%3 = 1, 2%3 = 2
    col = i % 3

    # Select the subplot position using row and column index
    ax = axes[row, col]

    # Plot histogram of the current numerical column
    sns.histplot(data=train_df[column], ax=ax, kde=True)

# Display all histograms on the screen
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(14, 10))

for i, column in enumerate(numerical_data):
    row = i // 3
    col = i % 3
    ax = axes[row, col]
    sns.kdeplot(data=train_df[column], ax=ax)

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(14, 10))

for i, column in enumerate(numerical_data):
    row = i // 3
    col = i % 3
    ax = axes[row, col]
    sns.boxplot(data=train_df[column], ax=ax)

fig.tight_layout()
plt.show()
```
