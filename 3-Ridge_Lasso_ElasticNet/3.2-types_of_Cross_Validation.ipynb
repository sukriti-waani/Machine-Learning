## 1Ô∏è‚É£ What is Cross Validation?

Cross Validation is a technique used to:

1. Train the model  
2. Tune hyperparameters  
3. Check model performance on unseen data

It helps in avoiding **overfitting**.


## 2Ô∏è‚É£ Dataset Split

Total Dataset ‚Üí 1000 records

- Training Data ‚Üí 70% (700)
- Testing Data ‚Üí 30% (300)

Training data is again split into:

- Train
- Validation

This internal split is called **Cross Validation**.


## 3Ô∏è‚É£ Leave One Out Cross Validation (LOOCV)

In LOOCV:

- If total samples = n  
- Training samples = n ‚àí 1  
- Validation samples = 1

This process is repeated **n times**.

Example:
If n = 500 ‚Üí 500 experiments are done.

### Advantages of LOOCV

- Uses almost all data for training
- Very accurate


### Disadvantages of LOOCV

- Very high computation cost
- Very slow for large datasets


## 4Ô∏è‚É£ K-Fold Cross Validation

Dataset is divided into **k equal parts (folds)**.

Example:
n = 500, k = 5  
Each fold size = 500 / 5 = 100


### Working of K-Fold CV

1. Use 4 folds for training  
2. Use 1 fold for validation  
3. Repeat for all folds  
4. Take **average accuracy**


## 5Ô∏è‚É£ Stratified K-Fold Cross Validation

Used mainly in **Classification** problems.

It keeps the **same class proportion** in all folds.

Example:
If dataset has:
- 60% class A
- 40% class B

Each fold will also have:
- 60% class A
- 40% class B


## 6Ô∏è‚É£ Time Series Cross Validation

Used when data depends on time.

Example:
- Product reviews by month
- Stock price prediction


### Time Series CV Rule

Past data is used to predict **future data**.

Training:
Day 1 ‚Üí Day 4

Validation:
Day 5 ‚Üí Day N


## 7Ô∏è‚É£ Overfitting Detection

- High Training Accuracy  
- Low Validation Accuracy  

This means the model is **Overfitting**.


## 8Ô∏è‚É£ Comparison Summary

LOOCV:
- High accuracy
- Very slow

K-Fold:
- Balanced approach
- Most commonly used

Stratified K-Fold:
- Best for imbalanced classes

Time Series CV:
- Used when order of data matters


## Summary

- Cross Validation improves model reliability
- LOOCV ‚Üí n experiments
- K-Fold ‚Üí average of k accuracies
- Stratified CV keeps class ratio
- Time Series CV respects time order
