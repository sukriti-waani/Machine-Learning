# üìò K-Nearest Neighbour (KNN)

---

## üîπ What is KNN?

**K-Nearest Neighbour (KNN)** is a **supervised machine learning algorithm** used for:

1. **Classification**
2. **Regression**

It works on the principle of **similarity** ‚Äì  
*A data point is classified based on its nearest neighbors.*

---

## üîπ Types of KNN

| Type | Output |
|------|--------|
| KNN Classification | Categorical (0 / 1, Yes / No, Spam / Not Spam) |
| KNN Regression | Continuous (House Price, Salary, Marks) |

---

## üîπ KNN for Classification

We use features like:

| f1 | f2 | y (Binary Category) |
|----|----|---------------------|
| ‚Äî | ‚Äî | 0 |
| ‚Äî | ‚Äî | 1 |

Example shown in diagram:  
- If majority neighbors are **0 ‚Üí Output = 0**  
- If majority neighbors are **1 ‚Üí Output = 1**

---

## üîπ Steps in KNN Classification

1. **Initialize K value**  
   - K > 0  
   - K = 1, 2, 3, 4, 5 ‚Ä¶ (Hyperparameter)

2. **Find the K Nearest Neighbours** for the test data.

3. **Count the categories** among K neighbours:
   - How many belong to **0 class**
   - How many belong to **1 class**

4. **Majority Voting ‚Üí Final Output**

### 1Ô∏è‚É£ Euclidean Distance

Distance between two points (x‚ÇÅ, y‚ÇÅ) and (x‚ÇÇ, y‚ÇÇ) is:

‚àö[(x‚ÇÇ ‚àí x‚ÇÅ)¬≤ + (y‚ÇÇ ‚àí y‚ÇÅ)¬≤]

Used when data is continuous.

---

### 2Ô∏è‚É£ Manhattan Distance

Distance between two points is:

|x‚ÇÇ ‚àí x‚ÇÅ| + |y‚ÇÇ ‚àí y‚ÇÅ|

Used when movement is grid-based (city block distance).

---

## üîπ KNN for Regression

Used when output is **continuous**.

Example:  
Predicting **House Price** based on **House Size**.

### Output Formula

Output = Average of K nearest neighbours

It takes the **average of the K nearest neighbors**.

---

## üîπ Time Complexity of KNN

O(n)

Because distance must be calculated from the test point to **every training point**.

---

## üîπ Optimization Techniques

To reduce time complexity:

| Technique |
|---------|
| KD-Tree |
| Ball Tree |
| Binary Tree |

These structures make KNN faster for large datasets.

---

## üìù Summary

- KNN is a **lazy learning algorithm** (no training phase).  
- It is used for **Classification and Regression**.  
- Works by finding **K nearest neighbours** using distance metrics.  
- Uses **majority voting** for classification and **average** for regression.  
- Time complexity is **O(n)** but can be optimized using **KD-Tree / Ball Tree**.
