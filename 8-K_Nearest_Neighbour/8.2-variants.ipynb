## üîπ Why Variants of KNN?

In normal KNN, we calculate distance from the test point to **every training point**.

So, **Time Complexity = O(n)**  
For large datasets (millions of points), this becomes very slow.

To optimize this, we use:

1. **KD Tree**  
2. **Ball Tree**

---

## 1Ô∏è‚É£ KD Tree (K-Dimensional Tree)

KD Tree is a **binary tree** that divides data based on feature dimensions.

---

### üî∏ Construction of KD Tree

1. Choose one feature (f‚ÇÅ or f‚ÇÇ).
2. Find the **median value** of that feature.
3. Split the data into left and right parts.
4. Repeat the process alternately for each dimension.

Example shown in diagram:

| f‚ÇÅ | f‚ÇÇ |
|----|----|
| 7 | 2 |
| 5 | 4 |
| 9 | 6 |
| 2 | 3 |
| 4 | 7 |
| 8 | 1 |

- Median of f‚ÇÅ ‚Üí root node (7,2)  
- Left child ‚Üí (5,4)  
- Right child ‚Üí (9,6)

---

### üî∏ Searching in KD Tree

- Traverse tree like **binary search**.
- Use **backtracking** when required.
- This avoids checking all points.

---

### üî∏ KD Tree Advantage

- Reduces searching time.
- Improves KNN performance.

---

## 2Ô∏è‚É£ Ball Tree

Ball Tree groups data points into **balls (clusters)**.

Each node represents:

- A **center point**
- A **radius** covering nearby points

---

### üî∏ Construction of Ball Tree

1. Divide dataset into groups.
2. Each group is enclosed inside a **ball (circle)**.
3. Balls are recursively split into smaller balls.

---

### üî∏ Searching in Ball Tree

- Distance is calculated between **test point and ball center**.
- If the ball is too far, it is ignored.

---

### üî∏ Ball Tree Advantage

- Works better for **high dimensional data**.
- Faster than KD Tree in large feature spaces.

---

## üîπ Comparison

| Feature | KD Tree | Ball Tree |
|--------|---------|-----------|
| Structure | Binary Tree | Hierarchical Balls |
| Best for | Low dimension data | High dimension data |
| Speed | Fast | Faster in complex data |

---

## üìù Exam-Ready Summary

- Normal KNN is slow for large datasets.
- **KD Tree** and **Ball Tree** optimize KNN.
- KD Tree splits data using medians.
- Ball Tree clusters data inside spherical regions.
- Both reduce time complexity and improve performance.
