## 1Ô∏è‚É£ What is Na√Øve Bayes?

Na√Øve Bayes is a **probability-based classification algorithm**.

It is based on:
1. Probability theory  
2. Bayes Theorem  
3. Assumption that all input features are **independent**.

---

## 2Ô∏è‚É£ Independent and Dependent Events

### Independent Events
Two events are independent if one event does not affect the other.

Example: Rolling a dice  
P(1) = 1/6 , P(2) = 1/6 , P(3) = 1/6

---

### Dependent Events

If the first event affects the second event.

Example:  
Removing an orange marble, then a yellow marble.

P(O and Y) = P(O) * P(Y|O)

Where  
P(Y|O) means probability of Y after O has happened.

---

## 3Ô∏è‚É£ Conditional Probability Formula

P(A and B) = P(A) * P(B|A)

---

## 4Ô∏è‚É£ Bayes Theorem

Bayes theorem formula:

P(A|B) = [ P(A) * P(B|A) ] / P(B)

Where:
- P(A|B) = Probability of A given B has occurred  
- P(A) = Probability of A  
- P(B) = Probability of B  
- P(B|A) = Probability of B given A has occurred  

---

## 5Ô∏è‚É£ Bayes Theorem for Classification

For input features x1, x2, x3 and output y:

P(y | x1, x2, x3) = [ P(y) * P(x1, x2, x3 | y) ] / P(x1, x2, x3)

---

## 6Ô∏è‚É£ Na√Øve Bayes Assumption

All features are independent:

P(x1, x2, x3 | y) = P(x1|y) * P(x2|y) * P(x3|y)

So final formula becomes:

P(y | x1, x2, x3) =
[ P(y) * P(x1|y) * P(x2|y) * P(x3|y) ] / P(x1) * P(x2) * P(x3)

Denominator is constant, so we ignore it while comparing.

---

## 7Ô∏è‚É£ Classification Rule

Calculate for each class:

P(Yes | x1, x2, x3)  
P(No  | x1, x2, x3)

The class with **higher probability is selected**.

---

## 8Ô∏è‚É£ Example ‚Äì Play Tennis Dataset

Given input:  
Outlook = Sunny  
Temperature = Hot  

We calculate:

P(Yes | Sunny, Hot)  
P(No | Sunny, Hot)

If:

P(No | Sunny, Hot) > P(Yes | Sunny, Hot)

Then prediction = **No (Not playing tennis)**

---

## üìù Summary

- Na√Øve Bayes uses Bayes Theorem.
- Assumes all features are independent.
- Class with maximum probability is chosen.
